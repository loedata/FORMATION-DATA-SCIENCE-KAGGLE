```python
# modules we'll use
import pandas as pd
import numpy as np

# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

# set seed for reproducibility
np.random.seed(0)
```

```python
# L'une des raisons pour lesquelles il est facile de confondre mise à 
# l'échelle et normalisation est que les termes sont parfois utilisés de 
# manière interchangeable et, pour rendre la chose encore plus confuse, 
# ils sont très similaires ! Dans les deux cas, vous transformez les 
# valeurs des variables numériques de manière à ce que les points de 
# données transformés aient des propriétés spécifiques utiles. 
# C'est la différence :
# dans la mise à l'échelle, vous modifiez l'étendue de vos données, 
# tandis que
# dans la normalisation, vous changez la forme de la distribution de vos données.
```

## 9.2.1.1. SCALING

```python
# Cela signifie que vous transformez vos données de manière à ce qu'elles
# s'inscrivent dans une échelle spécifique, comme 0-100 ou 0-1. 
# Vous voulez mettre les données à l'échelle lorsque vous utilisez des 
# méthodes basées sur des mesures de l'éloignement des points de données, 
# comme les machines à vecteur support (SVM) ou les voisins les plus
# proches (KNN). 
# Avec ces algorithmes, un changement de "1" dans une caractéristique 
# numérique quelconque se voit accorder la même importance.
```

```python
# generate 1000 data points randomly drawn from an exponential distribution
original_data = np.random.exponential(size=1000)

# mix-max scale the data between 0 and 1
scaled_data = minmax_scaling(original_data, columns=[0])

# plot both together to compare
fig, ax = plt.subplots(1,2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")

plt.show()
```

```python
# En mettant vos variables à l'échelle, vous pouvez aider à comparer 
# différentes variables sur un pied d'égalité.
```

## 9.2.1.2. NORMALIZATION

```python
# La mise à l'échelle ne fait que modifier la portée de vos données.
# La normalisation est une transformation plus radicale.
# Le but de la normalisation est de modifier vos observations afin 
# qu'elles puissent être décrites comme une distribution normale
```

```python
# Distribution normale : Aussi connue sous le nom de "courbe en cloche", 
# il s'agit d'une distribution statistique spécifique où des observations 
# à peu près égales se situent au-dessus et au-dessous de la moyenne, 
# la moyenne et la médiane sont les mêmes, et il y a plus d'observations 
# plus proches de la moyenne. La distribution normale est également connue 
# sous le nom de distribution gaussienne.
```

```python
# En général, vous normaliserez vos données si vous utilisez une technique 
# d'apprentissage automatique ou statistique qui suppose que vos données 
# sont distribuées normalement. L'analyse discriminante linéaire (LDA) et 
# la loi de Bayes naïve gaussienne en sont des exemples. 
# (Conseil du pro : toute méthode dont le nom comporte "Gaussien" suppose 
# probablement la normalité).
```

```python
# normalize the exponential data with boxcox
normalized_data = stats.boxcox(original_data)

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_data[0], ax=ax[1])
ax[1].set_title("Normalized data")
```

## 9.2.2.2. Practice normalization

```python
# get the index of all positive pledges (Box-Cox only takes positive values)
index_of_positive_pledges = kickstarters_2017.usd_pledged_real > 0

# get only positive pledges (using their indexes)
positive_pledges = kickstarters_2017.usd_pledged_real.loc[index_of_positive_pledges]

# normalize the pledges (w/ Box-Cox)
normalized_pledges = pd.Series(stats.boxcox(positive_pledges)[0], 
                               name='usd_pledged_real', index=positive_pledges.index)

# plot both together to compare
fig, ax=plt.subplots(1,2,figsize=(15,3))
sns.distplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_pledges, ax=ax[1])
ax[1].set_title("Normalized data")
plt.show()
```

```python
print('Original data\nPreview:\n', positive_pledges.head())
print('Minimum value:', float(positive_pledges.min()),
      '\nMaximum value:', float(positive_pledges.max()))
print('_'*30)

print('\nNormalized data\nPreview:\n', normalized_pledges.head())
print('Minimum value:', float(normalized_pledges.min()),
      '\nMaximum value:', float(normalized_pledges.max()))
```
