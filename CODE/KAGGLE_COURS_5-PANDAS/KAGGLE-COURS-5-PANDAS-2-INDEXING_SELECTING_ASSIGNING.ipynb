## INDEXING IN PANDAS

```python
# les accesseurs propres de Pandas sont loc et iloc
# à utiliser pour les opérations plus compliquées
```

```python
# l'indexation est basée sur 2 paradigmes :
# - INDEX-BASED SELECTION : sélectionner des données en fonction de leur 
#                           position numérique dans les données
# - LABEL-BASED SELECTION : sélection des données en fonction de leur
#                           valeur
```

## INDEX-BASED SELECTION : ILOC

```python
# pour récupérer la première ligne du DataFrame
reviews.iloc[0]
```

```python
# pour loc et iloc on passe toujours en premier lignes, en second colonnes
# => facile de récupérer les lignes et difficile de récupérer les colonnes
```

```python
# pour récupérer toutes les lignes de la première colonne avec iloc :
reviews.iloc[:, 0]
```

```python
# pour récupérer les 3 1ère lignes de la première colonne avec iloc :
reviews.iloc[:3,0]
```

```python
# pour récupérer les 2 1ères lignes de la première colonne avec iloc :
reviews.iloc[1:3,0]
```

```python
# on peut aussi passer une liste d'index de lignes
reviews.iloc[[1, 2], 0]
```

```python
# on peut aussi utiliser des nombres négatifs pour partir de la fin de la
# ligne
# les 5 dernières lignes pour toutes les colonnes
reviews.iloc[-5:]
```

## LABEL-BASED SELECTION : LOC

```python
# basé sur les valeurs : ligne puis colonne
```

```python
# pour récupérer la valeur de la première ligne 1ère colonne avec loc
# première ligne donc nom de l'index
# première colonne donc étiquette ou nom de la colonne
reviews.loc[0, 'country']
```

```python
# iloc est conceptuellement plus simple que loc car il ignore les indices 
# de l'ensemble de données. Lorsque nous utilisons iloc, nous traitons 
# l'ensemble de données comme une grande matrice (une liste de listes), 
# dans laquelle nous devons indexer par position. loc, en revanche, 
# utilise les informations contenues dans les indices pour faire son 
# travail. 
# Puisque votre ensemble de données a généralement des indices 
# significatifs, il est généralement plus facile de faire les choses en 
# utilisant loc à la place.
# Par exemple, voici une opération qui est beaucoup plus facile en 
# utilisant loc :
# toutes les lignes des colonnes taster_name, taster_twitter_handle et
# points
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

## CHOOSING BETWEEN LOC AND ILOC

```python
# iloc utilise le schéma d'indexation Python stdlib, où le premier élément 
# de la plage est inclus et le dernier exclu. 
# Ainsi, 0:10 sélectionnera les entrées 0, ..., 9. 
```

```python
# loc, quant à lui, indexe inclusivement. 
# Ainsi, 0:10 sélectionnera les entrées 0, ..., 10.
```

```python
# Pourquoi ce changement? 
# N'oubliez pas que loc peut indexer n'importe quel type stdlib: chaînes, 
# par exemple. 
# Si nous avons un DataFrame avec des valeurs d'index Apples, ..., Potatoes
# ..., et que nous voulons sélectionner "tous les choix de fruits 
# alphabétiques entre Apples, ..., Potatoet", alors il est beaucoup plus 
# pratique d'indexer df.loc [' Apples ':' Potatoes '] que pour indexer 
# quelque chose comme df.loc [' Apples ',' Potatoet] (t venant après s 
# dans l'alphabet).
```

```python
# Ceci est particulièrement déroutant lorsque l'index DataFrame est une 
# simple liste numérique, par ex. 0, ..., 1000. 
# Dans ce cas, df.iloc [0: 1000] retournera 1000 entrées, 
# tandis que df.loc [0: 1000] retournera 1001 d'entre elles! 
# Pour obtenir 1000 éléments en utilisant loc, vous devrez aller un plus 
# bas et demander df.loc [0: 999].
```

## CONDITIONAL SELECTION

```python
# sélectionner tous les vins d'italie ==> masque
reviews.country == 'Italy'
# c'est une Serie de Pandas, un masque contenant True ou False
```

```python
vins_italien = reviews.loc[reviews.country == 'Italy']
vins_italien
# vins_italien est une Serie de Pandas
```

```python
# pour voir les vins italiens avec un score >= 90
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```
