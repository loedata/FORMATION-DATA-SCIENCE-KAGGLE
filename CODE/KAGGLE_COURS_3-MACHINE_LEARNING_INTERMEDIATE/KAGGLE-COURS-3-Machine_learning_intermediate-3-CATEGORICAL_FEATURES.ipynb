## CHARGEMENT DES DATASETS FULL ET TEST

```python
# Read the data
# Chargement des datasets
dataset_input_path = 'C:/Users/PC Maison/4-KAGGLE\KAGGLE_DEV/KAGGLE_COURS_3-MACHINE_LEARNING_INTERMEDIATE/'
X = pd.read_csv(dataset_input_path + 'home-data-for-ml-course/input/train.csv', index_col='Id') 
X_test = pd.read_csv(dataset_input_path + 'home-data-for-ml-course/input/test.csv', index_col='Id')
```

## DEFINITION DES FEATURES ET DE LA TARGET

```python
# Remove rows with missing target, separate target from predictors
X.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X.SalePrice
X.drop(['SalePrice'], axis=1, inplace=True)

# To keep things simple, we'll drop columns with missing values
cols_with_missing = [col for col in X.columns if X[col].isnull().any()] 
X.drop(cols_with_missing, axis=1, inplace=True)
```

```python
X_test.drop(cols_with_missing, axis=1, inplace=True)
```

##### DECOUPE DATASET EN TRAIN SET ET VAL SET

```python
from sklearn.model_selection import train_test_split

# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y,
                                                      train_size=0.8, 
                                                      test_size=0.2,
                                                      random_state=0)
```

```python
X_train.head()
```

## MODELISATION AND SCORING

### Fonction de scoring : score_dataset()

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

# function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    # définition du model
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    # Entrainement sur le TRAIN SET
    model.fit(X_train, y_train)
    # Validation sur le VALID SET
    y_pred = model.predict(X_valid)
    # retourne la mAE entre les y_val et y_pred
    return model, mean_absolute_error(y_valid, y_pred)
```

## APPROCHE n°1 : SUPPRESSION DES COLONNES AVEC CATEGORICAL DATA

```python
# Fill in the lines below: drop columns in training and validation data
drop_X_train = X_train.select_dtypes(exclude='object')
drop_X_valid = X_valid.select_dtypes(exclude='object')
```

## SCORING

```python
print("MAE from Approach 1 (Drop categorical variables):")
model, MAE = score_dataset(drop_X_train, drop_X_valid, y_train, y_valid)
# print(score_dataset(drop_X_train, drop_X_valid, y_train, y_valid))
print(MAE)
```

## APPROCHE n°2 : LABEL ENCODING DES CATEGORICAL FEATURES

```python
print("Unique values in 'Condition2' column in training data:", 
      X_train['Condition2'].unique())
print("\nUnique values in 'Condition2' column in validation data:", 
      X_valid['Condition2'].unique())
```

```python
# ATTENTION toujours avoir les mêmes catégories dans X_train et X_Valid
# lors du découpage on peut ne trouver que des catégories dans un des 2
# set
# All categorical columns
object_cols = [col for col in X_train.columns if X_train[col].dtype == "object"]

# Columns that can be safely label encoded
good_label_cols = [col for col in object_cols if 
                   set(X_train[col]) == set(X_valid[col])]
        
# Problematic columns that will be dropped from the dataset
bad_label_cols = list(set(object_cols)-set(good_label_cols))
        
print('Categorical columns that will be label encoded:', good_label_cols)
print('\n Categorical columns that will be dropped from the dataset:', bad_label_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

# Drop categorical columns that will not be encoded
label_X_train = X_train.drop(bad_label_cols, axis=1)
label_X_valid = X_valid.drop(bad_label_cols, axis=1)

# Apply label encoder 
label_encoder= LabelEncoder()
# boucler sur les colonnes object = categorical feature 
for col in good_label_cols:
    # pour entrainer et tranfomer chaque colonne du x_train
    label_X_train[col] = label_encoder.fit_transform(X_train[col])
    # et pour transformer le label_X_valid set de chaque colonne du X_valid
    label_X_valid[col] = label_encoder.transform(X_valid[col])
```

## SCORING

```python
print("MAE from Approach 2 (Label Encoding):")
model, MAE = score_dataset(label_X_train, label_X_valid, y_train, y_valid)
# print(score_dataset(label_X_train, label_X_valid, y_train, y_valid))
print(MAE)
```

### CHOIX DES CATEGORIES AVEC PEU DE DIFFERENTES VAL

```python
# on ne travaille que sur les colonnes objets ayant moins de 10
# catégories différentes les autres peuvent être supprimées
# Columns that will be one-hot encoded
low_cardinality_cols = [col for col in object_cols if X_train[col].nunique() < 10]

# Columns that will be dropped from the dataset
high_cardinality_cols = list(set(object_cols)-set(low_cardinality_cols))

print('Categorical columns that will be one-hot encoded:', low_cardinality_cols)
print('\nCategorical columns that will be dropped from the dataset:', high_cardinality_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Use as many lines of code as you need!
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[low_cardinality_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[low_cardinality_cols]))

# One-hot encoding removed index; put it back
OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(object_cols, axis=1)
num_X_valid = X_valid.drop(object_cols, axis=1)

# Add one-hot encoded columns to numerical features
OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)
```

## SCORING

```python
print("MAE from Approach 3 (One-Hot Encoding):") 
model, MAE = score_dataset(OH_X_train, OH_X_valid, y_train, y_valid)
# print(score_dataset(OH_X_train, OH_X_valid, y_train, y_valid))
print(MAE)
```
