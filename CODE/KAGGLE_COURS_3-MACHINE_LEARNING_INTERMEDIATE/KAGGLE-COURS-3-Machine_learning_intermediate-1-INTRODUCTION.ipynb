## DEFINITION DES FEATURES ET DE LA TARGET

```python
# Sélection de la target
y = X_full.SalePrice
# Sélection des features
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 
            'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
# Datset de travail ==> copies
X = X_full[features].copy()
X_test = X_test_full[features].copy()
```

## DECOUPE DATASET EN TRAIN SET ET VAL SET

```python
# Découpe du dataset en TRAIN SET et VAL SET
from sklearn.model_selection import train_test_split

# Découpe en 2 jeux : TRAIN SET ET VAL SET
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, 
                                                      test_size=0.2,
                                                      random_state=0)
```

```python
# Quelques données du TRAIN SET
X_train.head()
```

## MODELISATION

### Modélisation RandomForest avec tests différents paramètres

```python
from sklearn.ensemble import RandomForestRegressor

# Define the models
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='mae', 
                                random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, 
                                random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, 
                                random_state=0)

models = [model_1, model_2, model_3, model_4, model_5]
```

## SCORING

## Sélection du meilleur model avec métrique MAE

```python
from sklearn.metrics import mean_absolute_error

# Function for comparing different models
def score_model(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    # Entrainement du modèle sur TRAIN SET X et y
    model.fit(X_t, y_t)
    # Prédictions sur le VAL TEST X
    preds = model.predict(X_v)
    # Calcul de la MAE
    return mean_absolute_error(y_v, preds)

# Boucle sur les différents modèles pour afficher la MAE de chaque modèle
dict_mae = {}
for i in range(0, len(models)):
    mae = score_model(models[i])
    dict_mae['modele n° '+ str(i+1)] = mae
    print("Model %d MAE: %d" % (i+1, mae))
print(f'{dict_mae = }')
```

```python
# Trouver le meilleur modèle
def get_keywithmaxval(dico):
    """ a) create a list of the dict's keys and values; 
         b) return the key with the max value"""  
    v=list(dico.values()) 
    k=list(dico.keys())
    key_min = k[v.index(min(v))]
    value_min = dico[key_min]
    return key_min, value_min
```

### Trouver meilleur modèle

```python
best_model, best_mae = get_keywithmaxval(dict_mae)
print('Meilleur modèle : ', best_model)
print('Plus petite MAE : ', best_mae)
```
